Task 3: Outlier Detection

```python
# Prepare data for outlier detection
outlier_df = south_asia_df.copy()
```

```python
# Convert numeric columns
outlier_df['hdi'] = pd.to_numeric(outlier_df['hdi'], errors='coerce')
outlier_df['gross_inc_percap'] = pd.to_numeric(outlier_df['gross_inc_percap'], errors='coerce')
```

```python
# Filter for latest available year for each country
latest_year = outlier_df.groupby('country')['year'].max().reset_index()
latest_data = pd.merge(outlier_df, latest_year, on=['country', 'year'])
```

```python
# Drop missing values
latest_data = latest_data.dropna(subset=['hdi', 'gross_inc_percap'])
```

```python
# Function to detect outliers using IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound
```

```python
# Detect outliers in HDI
hdi_outliers, hdi_lower, hdi_upper = detect_outliers_iqr(latest_data, 'hdi')
print(f"HDI Outliers ({len(hdi_outliers)} countries):")
if len(hdi_outliers) > 0:
    print(hdi_outliers[['country', 'year', 'hdi']])
print()
```

```python
# Detect outliers in GNI per Capita
gni_outliers, gni_lower, gni_upper = detect_outliers_iqr(latest_data, 'gross_inc_percap')
print(f"GNI per Capita Outliers ({len(gni_outliers)} countries):")
if len(gni_outliers) > 0:
    print(gni_outliers[['country', 'year', 'gross_inc_percap']])
print()
```

```python
# Combine outliers
all_outliers = pd.concat([hdi_outliers, gni_outliers]).drop_duplicates()
latest_data['is_outlier'] = latest_data['country'].isin(all_outliers['country'])
```

```python
# Create scatter plot with outliers highlighted
plt.figure(figsize=(10, 6))

# Plot non-outliers
normal_data = latest_data[~latest_data['is_outlier']]
plt.scatter(normal_data['gross_inc_percap'], normal_data['hdi'],
            alpha=0.6, label='Normal', color='blue')

# Plot outliers
if len(all_outliers) > 0:
    outlier_data = latest_data[latest_data['is_outlier']]
    plt.scatter(outlier_data['gross_inc_percap'], outlier_data['hdi'],
                alpha=0.8, label='Outliers', color='red', s=100, edgecolors='black')

    # Annotate outlier countries
    for idx, row in outlier_data.iterrows():
        plt.annotate(row['country'],
                    (row['gross_inc_percap'], row['hdi']),
                    xytext=(5, 5), textcoords='offset points')

plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.title('GNI per Capita vs HDI - South Asia (Outliers Highlighted)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Discuss outliers
print("Outlier Analysis:")
if len(all_outliers) > 0:
    for country in all_outliers['country'].unique():
        country_data = all_outliers[all_outliers['country'] == country].iloc[0]
        print(f"- {country}: HDI={country_data['hdi']:.3f}, GNI={country_data['gross_inc_percap']:.3f}")
        print(f"  May stand out due to unique economic structure, natural resources,")
        print(f"  or development patterns different from regional neighbors.")
else:
    print("No outliers detected in South Asian HDI or GNI per Capita data.")
```

Task 5: HDI Disparity

```python
# Calculate HDI range and coefficient of variation for each region
def calculate_range_cv(data, column):
    valid_data = data[column].dropna()
    data_range = valid_data.max() - valid_data.min()
    cv = valid_data.std() / valid_data.mean() if valid_data.mean() != 0 else np.nan
    return data_range, cv
```

```python
# Calculate for South Asia
sa_range, sa_cv = calculate_range_cv(south_asia_subset, 'hdi')
```

```python
# Calculate for Middle East
me_range, me_cv = calculate_range_cv(middle_east_subset, 'hdi')
```

Task 7: Outlier Detection

```python
# Function to detect outliers using IQR
def detect_outliers_iqr(data, column):
    if column not in data.columns:
        return pd.DataFrame(), 0, 0

    data[column] = pd.to_numeric(data[column], errors='coerce')
    clean_data = data[column].dropna()

    if len(clean_data) < 4:  # Need at least 4 points for IQR
        return pd.DataFrame(), 0, 0

    Q1 = clean_data.quantile(0.25)
    Q3 = clean_data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound
```

```python
# Prepare latest year data for each country in each region
def prepare_latest_year_data(df):
    # Get latest year for each country
    latest_years = df.groupby('country')['year'].max().reset_index()

    # Merge to get latest data for each country
    latest_data = pd.merge(df, latest_years, on=['country', 'year'])

    # Clean numeric columns
    if 'hdi' in latest_data.columns:
        latest_data['hdi'] = pd.to_numeric(latest_data['hdi'], errors='coerce')
    if 'gross_inc_percap' in latest_data.columns:
        latest_data['gross_inc_percap'] = pd.to_numeric(latest_data['gross_inc_percap'], errors='coerce')

    return latest_data.dropna(subset=['hdi', 'gross_inc_percap'])
```

```python
# Prepare data
sa_latest = prepare_latest_year_data(south_asia_subset)
me_latest = prepare_latest_year_data(middle_east_subset)
```

```python
# Detect outliers
print("Outlier Detection Results:")

# South Asia outliers
sa_hdi_outliers, sa_hdi_lower, sa_hdi_upper = detect_outliers_iqr(sa_latest, 'hdi')
sa_gni_outliers, sa_gni_lower, sa_gni_upper = detect_outliers_iqr(sa_latest, 'gross_inc_percap')

print(f"\nSouth Asia:")
print(f"HDI outliers: {len(sa_hdi_outliers)} countries")
if len(sa_hdi_outliers) > 0:
    print(sa_hdi_outliers[['country', 'hdi']])
print(f"GNI outliers: {len(sa_gni_outliers)} countries")
if len(sa_gni_outliers) > 0:
    print(sa_gni_outliers[['country', 'gross_inc_percap']])
```

```python
# Detect outliers
print("Outlier Detection Results:")

# Middle East outliers
me_hdi_outliers, me_hdi_lower, me_hdi_upper = detect_outliers_iqr(me_latest, 'hdi')
me_gni_outliers, me_gni_lower, me_gni_upper = detect_outliers_iqr(me_latest, 'gross_inc_percap')

print(f"\nMiddle East:")
print(f"HDI outliers: {len(me_hdi_outliers)} countries")
if len(me_hdi_outliers) > 0:
    print(me_hdi_outliers[['country', 'hdi']])
print(f"GNI outliers: {len(me_gni_outliers)} countries")
if len(me_gni_outliers) > 0:
    print(me_gni_outliers[['country', 'gross_inc_percap']])
```
